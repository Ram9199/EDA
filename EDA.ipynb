<font size="21">2. Importing Libraries and Loading the Dataset</font>

<font size="6">In this section, we import the necessary libraries and load the dataset for initial exploration.</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('raw_house_data.csv')

print("Dataset Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nFirst few rows:")
print(df.head())
```

<font size="6">Explanation:

Libraries Used:
pandas: For data manipulation and analysis.
numpy: For numerical operations and calculations.
seaborn: For data visualization.
Loading the Dataset:
We load the dataset using pd.read_csv(). This step is crucial as it gives us a raw view of the data.
Initial Exploration:
df.info(): Provides a summary of the dataset, including the number of non-null entries and data types.
df.describe(): Gives descriptive statistics for numerical columns.
df.isnull().sum(): Identifies columns with missing values.
df.head(): Displays the first few rows of the dataset. </font>
<br>
<font size="21">3. Exploratory Data Analysis (EDA)</font>

<font size="6">We conduct an exploratory data analysis to understand the distribution of numerical features, identify correlations, and detect outliers.</font>

```python
print(df.dtypes)

numerical_df = df.select_dtypes(include=['float64', 'int64'])

numerical_df.hist(bins=20, figsize=(14, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.show()


plt.figure(figsize=(12, 8))
corr_matrix = numerical_df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix', fontsize=16)
plt.show()

plt.figure(figsize=(14, 10))
sns.boxplot(data=numerical_df, palette='Set3')
plt.title('Box Plot of Numerical Features', fontsize=16)
plt.xticks(rotation=90) 
plt.show()


categorical_features = df.select_dtypes(include=['object']).columns

for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(y=feature, data=df, palette='viridis')
    plt.title(f'Distribution of {feature}', fontsize=16)
    plt.show()
```

<font size="6">Explanation:

Data Types: df.dtypes helps us identify numerical and categorical columns.
Numerical Features:
Histograms: Show the distribution of numerical features.
Correlation Matrix: Visualizes the relationship between numerical features using a heatmap.
Box Plots: Identify outliers in numerical features.
Categorical Features:
We loop through categorical columns and use count plots to visualize the distribution of each category. </font>

<font size="21">4. Data Cleaning and Preprocessing</font>

<font size="6">This step involves cleaning the data by handling missing values, encoding categorical variables, scaling numerical features, and removing outliers.</font>

```python

df_cleaned = df.dropna()

df_cleaned = df.fillna({
    'taxes': df['taxes'].mean(),
    'kitchen_features': df['kitchen_features'].mode()[0]
})

df_encoded = pd.get_dummies(df_cleaned, drop_first=True)

numerical_columns = df_encoded.select_dtypes(include=['float64', 'int64']).columns
df_encoded[numerical_columns] = (df_encoded[numerical_columns] - df_encoded[numerical_columns].mean()) / df_encoded[numerical_columns].std()

z_scores = np.abs((df_encoded[numerical_columns] - df_encoded[numerical_columns].mean()) / df_encoded[numerical_columns].std())
outlier_mask = (z_scores > 3).any(axis=1)
num_outliers = outlier_mask.sum()
print(f"Number of outliers detected: {num_outliers}")
print(f"Total rows before removing outliers: {len(df_encoded)}")

df_no_outliers = df_encoded[~outlier_mask]

print(f"Total rows after removing outliers: {len(df_no_outliers)}")

print("\nCleaned Dataset Information:")
print(df_no_outliers.info())
print(df_no_outliers.describe())

df_no_outliers.to_csv('cleaned_dataset.csv', index=False)
```

<font size="6">Explanation:

Handling Missing Values:
Dropping rows with missing values and filling remaining ones using appropriate statistics (mean for taxes, mode for kitchen_features).
One-Hot Encoding: Converts categorical variables into binary columns.
Scaling Numerical Features: Manual standardization using numpy ensures consistency across numerical data.
Outlier Detection and Removal: Using manual Z-score calculations with numpy, we detect and remove outliers.
Final Dataset Information: The cleaned dataset is ready for further analysis or modeling. </font>
